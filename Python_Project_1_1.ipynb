```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("train.csv")
data
```

```python
# 1. Find the maximum fare paid by a survivor.
max_fare_survivor= data[data["Survived"] == 1]["Fare"].max()
print("The highest fare paid by a survivor is:", max_fare_survivor)

```

```python
 # 2. Identify which class had the highest percentage of survivors.
survival_by_class = data.groupby("Pclass")["Survived"].mean()
class_with_highest_survival_rate= survival_by_class.idxmax()
print("The class with the highest survival percentage is:",class_with_highest_survival_rate)

```

```python
# 3. Create a stacked bar plot of survival by gender and class.

survival_by_gender_class = data.groupby(["Sex", "Pclass", "Survived"]).size().unstack()
survival_by_gender_class.plot(kind="bar", stacked=True, figsize=(8, 5))
plt.title("Survival by Gender and Class")
plt.ylabel("Number of Passengers")
plt.xlabel("Gender and Class")
plt.show()
```

```python
# 4. Find the survival rate for passengers aged 60 and above.
survival_rate_60_above = data[data["Age"] >= 60]["Survived"].mean()
print("The survival rate for passengers aged 60 and above is:", survival_rate_60_above)
```

```python
# 5. Count the number of unique ticket numbers in the dataset.
unique_ticket_count= data["Ticket"].nunique()
print("The number of unique ticket numbers is:", unique_ticket_count)
```

```python
# 6. Create a box plot of fare by survival status.

sns.boxplot(x="Survived", y="Fare", data=data)
plt.title("Box Plot of Fare by Survival Status")
plt.xlabel("Survival Status (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()

```

```python
print(f"Maximum Fare Paid by a Survivor: {max_fare_survivor}")
print(f"Class with Highest Percentage of Survivors: {class_with_highest_survival_rate}")
print(f"Survival Rate for Passengers Aged 60 and Above: {survival_rate_60_above}")
print(f"Number of Unique Ticket Numbers: {unique_ticket_count}")
```
